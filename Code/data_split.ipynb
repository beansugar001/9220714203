```python
import os
import shutil
import random
```

```python
# 1. 路径设置 (根据你的截图结构)
# 原始训练集路径 (含有两个子文件夹的那个)
train_base_path = r'D:\毕业论文\代码部分\datasets\images\train\PlantVillage'
# 验证集存放路径
val_base_path = r'D:\毕业论文\代码部分\datasets\images\val'
```

```python
# 2. 定义划分比例 (20% 作为验证集)
split_ratio = 0.2
```

```python
def split_data():
    # 获取分类文件夹名 (Tomato_healthy, Tomato__Tomato_YellowLeaf__Curl_Virus)
    categories = [d for d in os.listdir(train_base_path) if os.path.isdir(os.path.join(train_base_path, d))]
    
    print(f"检测到类别: {categories}")

    for category in categories:
        # 源分类目录
        src_category_path = os.path.join(train_base_path, category)
        # 目标验证集分类目录
        target_val_path = os.path.join(val_base_path, category)
        
        # 如果验证集目录不存在则创建
        os.makedirs(target_val_path, exist_ok=True)
        
        # 读取所有图片文件
        images = [f for f in os.listdir(src_category_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        
        # 随机打乱图片列表
        random.shuffle(images)
        
        # 计算需要移动到验证集的图片数量
        val_count = int(len(images) * split_ratio)
        val_images = images[:val_count]
        
        print(f"正在处理 {category}: 总数 {len(images)}, 移动 {val_count} 张到验证集...")
        
        # 执行移动操作 (注意：这里使用 move 是为了避免数据重复，如果想保留备份可以改成 copy)
        for img_name in val_images:
            src_path = os.path.join(src_category_path, img_name)
            dst_path = os.path.join(target_val_path, img_name)
            shutil.move(src_path, dst_path)

    print("\n--- 数据划分任务完成！---")
```

```python
if __name__ == "__main__":
    split_data()
```
